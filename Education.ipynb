```python
df.drop(columns = {
    'formhub_photo_id',
    'gps',
    'survey_id',
    'facility_id'
}, inplace=True)
df.head()
```

```python
df.rename(columns={
    'facility_name': 'School_name',
    'facility_type_display': 'School_type',
    'improved_water_supply': 'Improved_water_supply',
    'improved_sanitation': 'Improved_sanitation',
    'chalkboard_each_classroom_yn': 'Chalkboard_in_classroom',
    'num_tchrs_with_nce': 'Nce_qualified_teachers',
    'num_tchr_full_time': 'Full_time_teachers',
    'num_students_total': 'Total_students',
    'num_classrms_total': 'Total_classrooms',
    'phcn_electricity': 'Electricity_access',
    'num_toilets_total': 'Total_toilets',
    'num_students_male': 'Male_students',
    'num_students_female': 'Female_students',
    'num_tchrs_male': 'Male_teachers',	
    'num_tchrs_female':	'Female_teachers',
    'date_of_survey': 'Survey_date',
    'education_type': 'Education_type',	
    'community': 'Community',	
    'ward': 'Ward',	
    'management': 'Management_type',	
    'sector': 'Sector',	
    'unique_lga': 'Unique_lga',	
    'latitude': 'Latitude',	
    'longitude': 'Longitude'
    
}, inplace=True)
df.head()
```

```python
df.dropna(subset=['School_name'], inplace=True)
df['School_name'].isna().sum()
```

```python
df.dropna(subset=['Unique_lga'], inplace=True)
df['Unique_lga'].isna().sum()
```

```python
df.isna().sum()
```

```python
bool_cols = ['Improved_water_supply', 'Improved_sanitation', 'Chalkboard_in_classroom', 'Electricity_access']
for col in bool_cols:
    df[col] = df[col].astype(str).str.strip().str.lower()
```

```python
df.head()
```

```python
for col in bool_cols:
    df[col] = df[col].fillna(df[col].mode())
df[bool_cols].dtypes
```

```python
bool_cols = ['Improved_water_supply', 'Improved_sanitation', 'Chalkboard_in_classroom', 'Electricity_access']
for col in bool_cols:
    df[col] = df[col].map({'true': True, 'false': False})
    df[col] = df[col].astype(bool)
```

```python
df[bool_cols].dtypes
```

```python
df.head()
```

```python
catg_cols = ['School_type', 'Education_type', 'Community', 'Ward', 'Management_type']
for col in catg_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
num_cols = ['Nce_qualified_teachers', 'Full_time_teachers', 'Total_students', 'Total_classrooms', 'Total_toilets', 'Male_students',
            'Female_students', 'Male_teachers', 'Female_teachers']
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
from dateutil import parser
def date_parser(date_str):
    try:
        return parser.parse(date_str, dayfirst=True)
    except Exception:
        return pd.NaT
df['Survey_date'] = df['Survey_date'].apply(date_parser)

df['Survey_date']

```

```python
df['School_type'] = df['School_type'].astype(str).str.strip().str.lower()
```

```python
df['School_type'].value_counts()
```

```python
df['School_type'] = df['School_type'].replace({
    'primary only': 'primary school only',
    'pre-primary and primary': 'pre-primary and primary school combined',
    'junior secondary only': 'junior secondary school only',
    'junior and senior secondary': 'junior and senior secondary school combined',
    'primary, junior, and senior secondary': 'primary, junior and senior secondary school combined',
    'primary and junior secondary': 'primary and junior secondary school combined'
})
df['School_type'].value_counts()
```

```python
df = df[df['School_type'] != "information not available / don't know"]
df['School_type'].value_counts()
```

```python
df['Education_type'].value_counts()
```

```python
df['Education_type'] = df['Education_type'].astype(str).str.strip().str.lower()
df['Education_type'].value_counts()
```

```python
df['Education_type'] = df['Education_type'].replace({
    'formal only': 'formal_only',
    'religious only': 'religious_only'
})
```

```python
df['Education_type'] = df['Education_type']. replace('dk', df['Education_type'].mode()[0])
df['Education_type'].value_counts()
```

```python
df['Management_type'].value_counts()
```

```python
df = df[df['Management_type'].str.strip().str.lower() != "none"]
df['Management_type'].value_counts()
```

```python
df['Unique_lga'] = df['Unique_lga'].astype(str).str.strip().str.lower()
df['Unique_lga'].value_counts()
```
